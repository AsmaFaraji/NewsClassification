```python
from sklearn.cluster import KMeans
clf = KMeans(n_clusters=300)
clf.fit(X)
centroid = clf.cluster_centers_
labels = clf.labels_
```

```python
clusters = {}
indices = []
for l in range(0,max(labels) + 1):
    data_points = []
    indices = [i for i, x in enumerate(labels) if x == l]
    for idx in indices:
        data_points.append(df.loc[idx].tolist()[0])
    clusters[l] = data_points
```
