```python
df = pd.read_csv('/Users/asma//ThirdUpdateNewsDataSet/wordVectors.csv')
```

```python
X = df.iloc[:, 1:]
if df.loc[4].tolist()[0][1:-1] == 'FBI':
    print 'yes'
```

```python
from sklearn.cluster import KMeans
clf = KMeans(n_clusters=500)
clf.fit(X)
centroid = clf.cluster_centers_
labels = clf.labels_
```

```python
clusters = {}
indices = []
for l in range(0, max(labels)+1):
    data_points = []
    indices = [i for i, x in enumerate(labels) if x == l]
    for idx in indices:
        data_points.append(df.loc[idx].tolist()[0][1:-1])
    clusters[l] = data_points

```

```python
col = range(0,500)
col  = ['label']+ col
ds = pd.DataFrame(columns=col)
ds.to_csv('mydata.csv',header=True,index=False, encoding='utf-8')
```
